```python
path = 'data/Video_Games_Sales_as_at_22_Dec_2016.csv'
df = pd.read_csv(path)
print(df.shape)
df.info()
```

```python
df = df.dropna()
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df['User_Score'] = df.User_Score.astype('float64')
df['Year_of_Release'] = df.Year_of_Release.astype('int64')
df['User_Count'] = df.User_Count.astype('int64')
df['Critic_Count'] = df.Critic_Count.astype('int64')
df.info()
```

```python
useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 'Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'Rating']
df[useful_cols].head()
```

```python
sales_df = df[[x for x in df.columns if 'Sales' in x] + ['Year_of_Release']]
sales_df
```

```python
summarized_sales_df = sales_df.groupby('Year_of_Release').sum()
summarized_sales_df
```

```python
summarized_sales_df.plot()
```

```python
summarized_sales_df.plot(kind='bar', rot=45)
```

```python
import seaborn as sns
```

```python
cols = ['Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']
sns_plot = sns.pairplot(df[cols])
sns_plot.savefig('pairplot.png')
```

```python
sns.displot(df.Critic_Score)
```

```python
top_platforms = df.Platform.value_counts().sort_values(ascending=False).head(5).index.values
sns.boxplot(y='Platform', x='Critic_Score', data=df[df.Platform.isin(top_platforms)], orient='h')
```

```python
platform_genre_sales = df.pivot_table(index='Platform', columns='Genre', values='Global_Sales', aggfunc=sum).fillna(0).applymap(float)
sns.heatmap(platform_genre_sales, annot=True, fmt='.1f', linewidths=.5)
```

```python
from plotly.offline import download_plotlyjs, plot, iplot
import plotly.graph_objects as go
```

```python
years_df = df.groupby('Year_of_Release')[['Global_Sales']].sum().join(df.groupby('Year_of_Release')[['Name']].count())
years_df.columns = ['Global_Sales', 'Number_of_Games']
years_df
```

```python
trace0 = go.Scatter(
    x=years_df.index,
    y=years_df.Global_Sales,
    name='Global Sales'
)

trace1 = go.Scatter(
    x=years_df.index,
    y=years_df.Number_of_Games,
    name='Number of Games released'
)

data = [trace0, trace1]
layout = {'title': 'Statistics of Videogames'}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
plot(fig, filename='years_stats.html', show_link=False)
```

```python
platforms_df = df.groupby('Platform')[['Global_Sales']].sum().join(df.groupby('Platform')[['Name']].count())
platforms_df.columns = ['Global_Sales', 'Number_of_Games']
platforms_df.sort_values('Global_Sales', ascending=False, inplace=True)

trace0 = go.Bar(
    x=platforms_df.index,
    y=platforms_df.Global_Sales,
    name='Global Sales'
)

trace1 = go.Bar(
    x=platforms_df.index,
    y=platforms_df.Number_of_Games,
    name='Number of games released'
)

data = [trace0, trace1]

layout = {'title': 'Share of platforms', 'xaxis': {'title': 'platform'}}

fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
data = []
for genre in df.Genre.unique():
    data.append(
        go.Box(y=df[df.Genre==genre].Critic_Score, name=genre)
    )

iplot(data, show_link=False)
```
