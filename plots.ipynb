```python
import numpy as np


rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))
print(rnd)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(rng, rnd, labels=['f', 'g', 'h'])
ax.set_title('Graphic name')
ax.legend(loc='upper left')
ax.set_ylabel('Randoms')
ax.set_xlim(xmin=rng[0], xmax=rng[-1])
fig.tight_layout()

plt.show()
          
```
